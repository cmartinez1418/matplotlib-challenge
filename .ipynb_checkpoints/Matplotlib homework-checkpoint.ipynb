```python
#Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


#merge data frames
mouse_study = pd.merge(study_results, mouse_metadata, on=["Mouse ID"])

mouse_study
```

```python
#number of mice
mouse_study["Mouse ID"].nunique()
```

```python
#getting duplicate mice by ID number
duplicate_mice = mouse_study.loc[mouse_study.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
duplicate_mice
```

```python
#info on duplicate mice
duplicate_mice_info = mouse_study[mouse_study["Mouse ID"].isin(duplicate_mice)]
duplicate_mice_info
```

```python
#getting rid of duplicate
mouse_study_clean = mouse_study[mouse_study["Mouse ID"].isin(duplicate_mice)==False]
mouse_study_clean
```

```python
mouse_study_clean["Mouse ID"].nunique()
```

Summary Statistics

```python
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean = mouse_study_clean["Tumor Volume (mm3)"].groupby(mouse_study_clean["Drug Regimen"]).mean()
median = mouse_study_clean["Tumor Volume (mm3)"].groupby(mouse_study_clean["Drug Regimen"]).median()
variance = mouse_study_clean["Tumor Volume (mm3)"].groupby(mouse_study_clean["Drug Regimen"]).var()
std_dev = mouse_study_clean["Tumor Volume (mm3)"].groupby(mouse_study_clean["Drug Regimen"]).std()
std_error = mouse_study_clean["Tumor Volume (mm3)"].groupby(mouse_study_clean["Drug Regimen"]).sem()

summary_df = pd.DataFrame({"Mean Tumor Volume": mean, "Median Tumor Volume":median, "Tumor Volume Variance": variance, 
             "Tumor Volume Std. Dev.": std_dev, "Tumor Volume Std. Err.": std_error})
summary_df
```

```python
#summary using aggregate method
summary_df_agg = mouse_study_clean.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std", "sem"])
summary_df_agg
```

Quartiles, Outliers, and Boxplots

```python
#final tumor volume of each mouse for Capomulin, Ramicane, Infubinol, and Ceftamin
#get the last timepoint for each mouse

last_timepoint = mouse_study_clean.groupby("Mouse ID")["Timepoint"].max()
last_timepoint_df = pd.DataFrame(last_timepoint)

mouse_study_last_timepoint_df = pd.merge(last_timepoint_df, mouse_study_clean, on=["Mouse ID", "Timepoint"])
mouse_study_last_timepoint_df
```

```python
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
tumor_volume = [] 

for drug in treatments:
    volumes = mouse_study_last_timepoint_df.loc[mouse_study_last_timepoint_df["Drug Regimen"]==drug]["Tumor Volume (mm3)"]
    tumor_volume.append(volumes)
    quartiles = volumes.quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = (volumes.loc[volumes > upper_bound]) | (volumes.loc[volumes < lower_bound])
    print(f"{drug}'s potential outliers: {outliers}")

```

```python
plt.boxplot(tumor_volume)
plt.xticks([1, 2, 3, 4], treatments)
plt.ylabel("Final Tumor Volum (mm3)")
```

Line and Scatter Plots

```python
#line plot for mouse t565 on capomulin
capomulin = mouse_study_clean.loc[(mouse_study_clean["Drug Regimen"]=="Capomulin")]
capomulin_df = pd.DataFrame(capomulin)
#capomulin_df.value_counts("Mouse ID")
t565_df = pd.DataFrame(capomulin_df.loc[(capomulin_df["Mouse ID"]=="t565")])
timepoint = t565_df["Timepoint"]
volume = t565_df["Tumor Volume (mm3)"]

plt.plot(timepoint, volume)
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin treatment of mouse t565")
```

```python
#scatter plot for average tumor volume and weight
average_tumor = pd.DataFrame(capomulin_df.groupby("Mouse ID")["Tumor Volume (mm3)"].mean())
weights_df = pd.DataFrame(capomulin_df.groupby("Mouse ID")["Weight (g)"].mean())
tumor_weight_df = pd.merge(average_tumor, weights_df, on=["Mouse ID"])
tumor_weight_df.plot(kind="scatter", x="Weight (g)", y="Tumor Volume (mm3)")
plt.ylabel("Average Tumor Volume (mm3)")
```

Correlation and Regression

```python
#calculate correlation coefficient and linear regression for scatter plot above
x_values = tumor_weight_df["Weight (g)"]
y_values = tumor_weight_df["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, "r-")
correlation = st.pearsonr(x_values, y_values)
print(f"The correlation between mouse weight and average tumor volume is {round(correlation[0],2)}.")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
```
