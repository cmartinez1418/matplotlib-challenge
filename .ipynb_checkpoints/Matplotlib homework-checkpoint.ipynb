```python
#Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


#merge data frames
mouse_study = pd.merge(study_results, mouse_metadata, on=["Mouse ID"])

mouse_study
```

```python
#number of mice
mouse_study["Mouse ID"].nunique()
```

```python
#getting duplicate mice by ID number
duplicate_mice = mouse_study.loc[mouse_study.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
duplicate_mice
```

```python
#info on duplicate mice
duplicate_mice_info = mouse_study[mouse_study["Mouse ID"].isin(duplicate_mice)]
duplicate_mice_info
```

```python
#getting rid of duplicate
mouse_study_clean = mouse_study[mouse_study["Mouse ID"].isin(duplicate_mice)==False]
mouse_study_clean
```

```python
mouse_study_clean["Mouse ID"].nunique()
```

Summary Statistics

```python
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean = mouse_study_clean["Tumor Volume (mm3)"].groupby(mouse_study_clean["Drug Regimen"]).mean()
median = mouse_study_clean["Tumor Volume (mm3)"].groupby(mouse_study_clean["Drug Regimen"]).median()
variance = mouse_study_clean["Tumor Volume (mm3)"].groupby(mouse_study_clean["Drug Regimen"]).var()
std_dev = mouse_study_clean["Tumor Volume (mm3)"].groupby(mouse_study_clean["Drug Regimen"]).std()
std_error = mouse_study_clean["Tumor Volume (mm3)"].groupby(mouse_study_clean["Drug Regimen"]).sem()

summary_df = pd.DataFrame({"Mean Tumor Volume": mean, "Median Tumor Volume":median, "Tumor Volume Variance": variance, 
             "Tumor Volume Std. Dev.": std_dev, "Tumor Volume Std. Err.": std_error})
summary_df
```

```python
#summary using aggregate method
summary_df_agg = mouse_study_clean.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std", "sem"])
summary_df_agg
```

Bar and Pie Charts

```python
#bar plot showing the total number of timpeoints for all mice tested for each drug regimen using pandas

counts_per_regimen = mouse_study_clean["Drug Regimen"].value_counts()
#counts_per_regimen

pandas_plot = counts_per_regimen.plot.bar(color="g")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Mice Tested")
```

```python
#bar plot showing the total number of timpeoints for all mice tested for each drug regimen using pandas
x_axis = counts_per_regimen.index.values
y_axis = counts_per_regimen.values

plt.bar(x_axis, y_axis, color="g", alpha=1, align="center")
plt.xticks(rotation="vertical")
plt.ylabel("Number of Mice Tested")
plt.xlabel("Drug Regimen")
```

```python
# pie chart showing male vs female mice using pandas
counts_by_sex = mouse_study_clean["Sex"].value_counts()

pandas_pie = counts_by_sex.plot.pie(colors=["lightblue", "orange"], autopct="%1.1f%%")
plt.ylabel("Sex")
```

```python
#pie chart showing male vs female using pyplot
mouse_study_clean["Sex"].value_counts()

labels = ["Male", "Female"]
colors = ["lightblue", "orange"]
sizes = [958, 922]
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%")
plt.ylabel("Sex")
```
